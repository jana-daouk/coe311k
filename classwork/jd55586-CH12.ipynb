```python
import numpy as np

def setup_tridiagonal_matrix(x, y):
    n = len(x)

    # spacing between x values
    h = np.diff(x)

    # create arrays for the tridiagonal system
    A = np.zeros(n - 2)  # subdiagonal
    B = np.zeros(n - 2)  # main diagonal
    C = np.zeros(n - 2)  # superdiagonal
    D = np.zeros(n - 2)  # right-hand side

    # fill in values for interior points
    for i in range(1, n - 1):
        A[i - 1] = h[i - 1]  # left spacing
        B[i - 1] = 2 * (h[i - 1] + h[i])  # center weight
        C[i - 1] = h[i]  # right spacing

        # difference in slopes (how much curve we need)
        D[i - 1] = 6 * (
            (y[i + 1] - y[i]) / h[i] # slope going right
            - (y[i] - y[i - 1]) / h[i - 1]   # slope coming from left
        )

    return A, B, C, D


# example data
x = np.array([0, 1, 2, 3, 4, 5])
y = np.array([0, 1, 0, 1, 0, 1])

# build system
A, B, C, D = setup_tridiagonal_matrix(x, y)

print("Subdiagonal (A):", A)
print("Main diagonal (B):", B)
print("Superdiagonal (C):", C)
print("Right-hand side (D):", D)
```

This code constructs the tridiagonal system used in cubic spline interpolation by computing the coefficients A, B, C and the right-hand side D. The values in D represent changes in slope between intervals, which determine the curvature of the spline.
