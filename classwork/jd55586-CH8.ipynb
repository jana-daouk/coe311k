```python
def is_diagonally_dominant(A):
    n = len(A)  # number of rows (assumes square matrix)

    # loop through each row
    for i in range(n):
        diag = abs(A[i][i])  # absolute value of the diagonal entry

        off_diag_sum = 0  # sum of absolute values of non-diagonal entries

        # loop through columns in row i
        for j in range(n):
            if j != i:  # skip the diagonal element
                off_diag_sum += abs(A[i][j])

        # check diagonal dominance condition for this row
        if diag < off_diag_sum:
            return False  # fails for this row then matrix is not diagonally dominant

    # if all rows satisfy the condition
    return True
```

```python
A = [
    [3, 1, 1],
    [2, 6, 1],
    [2, 2, 5]
]

print(A)
print("Is the matrix diagonally dominant?", is_diagonally_dominant(A))
```
