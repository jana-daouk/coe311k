```python
import random
import numpy as np
import matplotlib.pyplot as plt

# monte carlo function to approximate pi
def montecarlo_pi(number_of_shots):
    
    number_of_hits = 0
    
    for i in range(number_of_shots):
        x = random.uniform(-1, 1)
        y = random.uniform(-1, 1)
        
        # check if point is inside unit circle
        if (x*x + y*y) < 1:
            number_of_hits += 1
            
    return 4 * number_of_hits / number_of_shots
```

```python
import math 
# need to store error numbers somewhere to be able to plot them
absolute_errors = []
relative_errors = []

for i in range(100):
    calculated_pi = montecarlo_pi(1000)
    true_pi = np.pi
    
    true_error = abs(true_pi - calculated_pi)
    relative_error = abs((true_pi - calculated_pi) / true_pi)
    
    absolute_errors.append(true_error)
    relative_errors.append(relative_error)

```

```python
plt.plot(range(1, 101), absolute_errors)
plt.xlabel("Run number")
plt.ylabel("Absolute error")
plt.title("Absolute Error over 100 Monte Carlo Runs")
plt.show()
```

```python
plt.plot(range(1, 101), relative_errors)
plt.xlabel("Run number")
plt.ylabel("Relative error")
plt.title("Relative Error over 100 Monte Carlo Runs")
plt.show()
```
