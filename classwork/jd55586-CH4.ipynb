```python
import numpy as np

pi_sum = 0  # running sum of series

for n in range(5000):  # first 5000 terms
    
    # add the next Leibniz term
    pi_sum += ((-1)**n) / (2*n + 1)
    
    # every 100 terms check how bad it is 
    if (n + 1) % 100 == 0:
        
        calculated_pi = 4 * pi_sum  # Leibniz gives pi/4 so multiply by 4
        
        # relative error formula
        rel_error = abs((np.pi - calculated_pi) / np.pi)
        
        print("After", n+1, "terms:")
        print("Relative Error =", rel_error)
        print()
```
