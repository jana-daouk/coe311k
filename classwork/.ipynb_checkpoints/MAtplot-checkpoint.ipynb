```python
import matplotlib.pyplot as plt 
import numpy as np
```

```python
x = np.linspace(0, 2*np.pi, 50)
```

```python
x
```

```python
plt.plot(x, np.sin(x), x, "r--", 2*np.sin(x), "b-o")
plt.show()
```

```python
plt.subplot(2,1,1) #row, column, active area
plt.plot(x, np.sin(x), 'b')
plt.subplot(2,1,2)
plt.plot(x, np.cos(x), 'g')
plt.subplot(2,2,3)
plt.plot(x, np.tan(x), 'k')
```

```python
plt.plot(x, np.sin(x))
```

```python
plt.scatter(x, np.sin(x))
```
