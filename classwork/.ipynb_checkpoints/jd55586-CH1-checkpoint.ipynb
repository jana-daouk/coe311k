```python
import matplotlib.pyplot as plt
import numpy as np
x = np.array([1, 2, 3])

plt.figure()

plt.subplot(221)
plt.plot(x)
plt.scatter(x, x)
plt.bar(x, x)
plt.hist(x)

plt.show()
```
