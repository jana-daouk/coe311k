```python
import numpy as np
import matplotlib.pyplot as plt

# Define the function and interval
f = lambda x: np.sin(x)
a, b = 0, 2 * np.pi
N = 50

# Create x values
x = np.linspace(a, b, N)

# Step size
h = (b - a) / (N - 1)

# True derivative
true_derivative = np.cos(x)

# Central Difference Approximation
central_diff = (f(x + h) - f(x - h)) / (2 * h)

# Forward Difference Approximation
forward_diff = (f(x + h) - f(x)) / h

# Backward Difference Approximation
backward_diff = (f(x) - f(x - h)) / h

# Plot everything in 4 subplots
plt.figure(figsize=(12, 10))

# 1. Exact derivative
plt.subplot(2, 2, 1)
plt.plot(x, true_derivative, label='cos(x)')
plt.xlabel('x')
plt.ylabel("f'(x)")
plt.title("Exact Derivative")
plt.legend()

# 2. Forward difference
plt.subplot(2, 2, 2)
plt.plot(x, true_derivative, label='cos(x)')
plt.plot(x, forward_diff, 'o--', label='Forward Difference')
plt.xlabel('x')
plt.ylabel("f'(x)")
plt.title("Forward Difference")
plt.legend()

# 3. Backward difference
plt.subplot(2, 2, 3)
plt.plot(x, true_derivative, label='cos(x)')
plt.plot(x, backward_diff, 'o--', label='Backward Difference')
plt.xlabel('x')
plt.ylabel("f'(x)")
plt.title("Backward Difference")
plt.legend()

# 4. Central difference
plt.subplot(2, 2, 4)
plt.plot(x, true_derivative, label='cos(x)')
plt.plot(x, central_diff, 'o--', label='Central Difference')
plt.xlabel('x')
plt.ylabel("f'(x)")
plt.title("Central Difference")
plt.legend()

plt.tight_layout()
plt.show()
```

The central difference method provides the most accurate approximation, while forward and backward differences show larger errors.
