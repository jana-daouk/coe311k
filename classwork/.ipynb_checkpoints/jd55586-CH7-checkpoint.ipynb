```python
import numpy as np
import time

def jacobi(A, b, x0, tol=1e-6, max_iterations=10000, blowup=1e12):
    n = len(b)
    x = x0.copy().astype(float)

    D = np.diag(A)
    R = A - np.diagflat(D)

    for k in range(max_iterations):
        x_new = (b - R @ x) / D

        if np.linalg.norm(x_new - x, ord=np.inf) < tol:
            return x_new, (k + 1), True

        if (not np.all(np.isfinite(x_new))) or (np.linalg.norm(x_new, ord=np.inf) > blowup):
            return x_new, (k + 1), False

        x = x_new

    return x, max_iterations, False

# used code from prev challenge
# have blowup checker because it kept erroring out (looked up)
def gauss_seidel(A, b, x0, tol=1e-6, max_iterations=10000, blowup=1e12):
    n = len(b)
    x = x0.copy().astype(float)

    for k in range(max_iterations):
        x_new = x.copy()

        for i in range(n):
            s1 = sum(A[i][j] * x_new[j] for j in range(i))
            s2 = sum(A[i][j] * x[j] for j in range(i + 1, n))
            x_new[i] = (b[i] - s1 - s2) / A[i][i]

        if np.linalg.norm(x_new - x, ord=np.inf) < tol:
            return x_new, (k + 1), True

        if (not np.all(np.isfinite(x_new))) or (np.linalg.norm(x_new, ord=np.inf) > blowup):
            return x_new, (k + 1), False

        x = x_new

    return x, max_iterations, False


def run(case_name, A, b, tol=1e-6):
    print(f"\n{case_name}")
    x0 = np.zeros_like(b, dtype=float)

    t0 = time.perf_counter()
    x_j, it_j, ok_j = jacobi(A, b, x0, tol=tol)
    t_j = time.perf_counter() - t0

    t0 = time.perf_counter()
    x_gs, it_gs, ok_gs = gauss_seidel(A, b, x0, tol=tol)
    t_gs = time.perf_counter() - t0

    print("Jacobi:")
    print("  converged:", ok_j)
    print("  iterations:", it_j)
    print("  time (sec):", t_j)
    print("  x:", x_j)

    print("Gauss-Seidel:")
    print("  converged:", ok_gs)
    print("  iterations:", it_gs)
    print("  time (sec):", t_gs)
    print("  x:", x_gs)




# Case 1
A1 = np.array([
    [0.582745, 0.48    , 0.10    , 0.     , 0.     ],
    [0.48    , 1.044129, 0.46    , 0.10   , 0.     ],
    [0.10    , 0.46    , 1.10431 , 0.44   , 0.10   ],
    [0.      , 0.10    , 0.44    , 0.963889, 0.42  ],
    [0.      , 0.      , 0.10    , 0.42   , 0.522565]
], dtype=float)

b1 = np.array([1.162745, 2.084129, 2.20431, 1.923889, 1.042565], dtype=float)

# Case 2
A2 = np.array([[1, 1], [1, 1.0001]], dtype=float)
b2 = np.array([1, 1], dtype=float)

# Case 3 
A3 = np.array([[1, 1], [1, 1.0001]], dtype=float)
b3 = np.array([1, 1.0001], dtype=float)

run("Case 1: 5x5 system", A1, b1, tol=1e-6)
run("Case 2: A=[[1,1],[1,1.0001]], b=[1,1]", A2, b2, tol=1e-6)
run("Case 3: A=[[1,1],[1,1.0001]], b=[1,1.0001]", A3, b3, tol=1e-6)
```

For all tested cases, Gauss–Seidel converged in fewer iterations than Jacobi because it updates each component using the latest available values.
