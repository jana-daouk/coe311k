```python
import numpy as np
import matplotlib.pyplot as plt

def taylor_sin(x, n_terms):
    
    x = np.asarray(x, dtype=float)

    total = 0.0
    for k in range(n_terms):
        power = 2*k + 1
        term = ((-1)**k) * (x**power) / np.math.factorial(power)
        total += term

    return total

# x from 0 to 10 in steps of 0.1 looked this up lowkey
x_vals = np.arange(0, 10.0 + 1e-12, 0.1)

# exact sin(x)
exact = np.sin(x_vals)

# do 1, 2, 3, 4 terms
term_counts = [1, 2, 3, 4]
errors = {}

for n in term_counts:
    approx = taylor_sin(x_vals, n)
    # absolute error append to error
    errors[n] = np.abs(approx - exact)   

# plot truncation error for 1/2/3/4 terms for each x
plt.figure()
for n in term_counts:
    plt.plot(x_vals, errors[n], label=f"{n} term(s)")
plt.xlabel("x")
plt.ylabel("Absolute error")
plt.title("Taylor Series Approx Error for sin(x)")
plt.legend()
plt.show()

# discussion (I didn't know where else to put it)
print("\nDiscussion:")
print("As I add more terms, the error usually drops a lot for small x.")
print("But the error doesn’t decrease the same way for every x, it gets worse for bigger x when I only use 1 to 4 terms.")
print("So the truncated Taylor series works best near 0, and for larger x you’d need more terms to stay accurate.")
```
