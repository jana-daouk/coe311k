```python
import random
import numpy as np
import matplotlib.pyplot as plt

# monte carlo function to approximate pi
def montecarlo_pi(number_of_shots):
    
    number_of_hits = 0
    
    for i in range(number_of_shots):
        x = random.uniform(-1, 1)
        y = random.uniform(-1, 1)
        
        # check if point is inside unit circle
        if (x*x + y*y) < 1:
            number_of_hits += 1
            
    return 4 * number_of_hits / number_of_shots
```

```python
import numpy as np
import matplotlib.pyplot as plt

# number of samples from 10 to 1000 increasing by 20 each time
samples = range(10, 1001, 20)
errors = []

for n in samples:
    calculated_pi = montecarlo_pi(n)
    # formula for error from instructions
    errors.append(abs(np.pi - calculated_pi))

plt.plot(samples, errors)
plt.xlabel("Number of samples")
plt.ylabel("Absolute error")
plt.show()

#My comment on the graph: As the number of samples increases from 10 to 1000, the absolute error generally decreases despite random fluctuations.
```
